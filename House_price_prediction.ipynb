```python
X =hp[['number of bedrooms', 'number of bathrooms',
       'number of floors',
       'condition of the house',
       'Number of schools nearby']]
Y = hp['Price']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
bedroom_encoder = LabelEncoder()
bathroom_encoder = LabelEncoder()
condition_encoder = LabelEncoder()
floors_encoder = LabelEncoder()
schools_encoder = LabelEncoder()
```

```python
import warnings
warnings.filterwarnings('ignore')
bedroom_encoder.fit(pd.DataFrame(X['number of bedrooms']))
X['number of bedrooms'] = bedroom_encoder.transform(X['number of bedrooms'])

bathroom_encoder.fit(pd.DataFrame(X['number of bathrooms']))
X['number of bathrooms'] = bathroom_encoder.transform(X['number of bathrooms'])

floors_encoder.fit(pd.DataFrame(X['number of floors']))
X['number of floors'] = floors_encoder.transform(X['number of floors'])

condition_encoder.fit(pd.DataFrame(X['condition of the house']))
X['condition of the house'] = condition_encoder.transform(X['condition of the house'])


schools_encoder.fit(pd.DataFrame(X['Number of schools nearby']))
X['Number of schools nearby'] = schools_encoder.transform(X['Number of schools nearby'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
Y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
score = r2_score(Y_test,Y_predict)
mae = mean_absolute_error(Y_test,Y_predict)
mse = mean_squared_error(Y_test,Y_predict)
print("R2 score is ", score)
print(" Mean absolute error is ",mae)
print( " Mean squared erroe is ",mse)
```

```python
from flask import Flask
from flask import render_template
import joblib
from flask import request
```

```python
app = Flask(__name__)
model = joblib.load('House_pred_india.pkl')
@app.route('/')
def hello(name=None):
    return render_template('index1.html',name=name)

@app.route('/results1',methods=['GET','POST'])
def get_value():
    if request.method=="POST":
        bedrooms=int(request.form.get('bedrooms'))
        bathrooms=int(request.form.get('bathrooms'))
        floors=float(request.form.get('floors'))
        condition=int(request.form.get('condition'))
        schools=int(request.form.get('schools'))
        predict=model.predict([[bedrooms,bathrooms,floors,condition,schools]])
        print(predict[0])
        
    return render_template('results1.html',prediction=[predict[0]])
```

```python
if __name__=="__main__":
    app.run()
```
