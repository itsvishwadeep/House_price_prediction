```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
india_file_path = 'House_Price_India.csv'
house_data = pd.read_csv(india_file_path)
house_data.describe()
```

```python
house_data.shape
```

```python
house_data.head(5)
```

```python
house_data = house_data.dropna(axis=0)
```

```python
Y = house_data.Price
```

```python
house_features = ['number of bedrooms','number of bathrooms','living area','Lattitude','Longitude','Number of schools nearby','Distance from the airport']

                  
```

```python
X = house_data[house_features]
```

```python
X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
india_model = DecisionTreeRegressor(random_state=1)
india_model.fit(X,Y)
```

```python
print("Making predictions for the following 5 house:")
print(X.head())
print("The prediction are")
print(india_model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
predicted_home_prices = india_model.predict(X)
mean_absolute_error(Y,predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X,val_X,train_Y,val_Y = train_test_split(X,Y,random_state=0)
```

```python
india_model = DecisionTreeRegressor()
```

```python
india_model.fit(train_X,train_Y)
```

```python
val_prediction = india_model.predict(val_X)
print(mean_absolute_error(val_Y,val_prediction))
```
